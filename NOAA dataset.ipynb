```python
import boto3
from botocore import UNSIGNED
from botocore.config import Config
import json
```

```python
s3_client = boto3.client('s3', config=Config(signature_version=UNSIGNED))
```

```python
s3_contents = s3_client.list_objects_v2(Bucket='noaa-goes16')['Contents']
```

```python
print(json.dumps([c['Key'] for c in s3_contents], indent=4))
```
